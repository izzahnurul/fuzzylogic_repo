## C. Outlier

```python
# menghapus outliner menggunakan z-score untuk data numerical
from scipy import stats
```

```python
print(f'Jumlah baris sebelum menggunakan z-score (Outlier): {len(df)}')

filtered_entries = np.array([True] * len(df))

for col in nums:
    zscore = abs(stats.zscore(df[col])) # hitung absolute z-scorenya
    filtered_entries = (zscore < 3) & filtered_entries # keep yang kurang dari 3 absolute z-scorenya
    
df_bersih = df[filtered_entries] # filter, cuma ambil yang z-scorenya dibawah 3

print(f'Jumlah baris setelah menggunakan z-score (Outlier): {len(df_bersih)}')
```

Setelah menggunakan z-score untuk menghilangkan outlier, jumlah baris sebelum z-score dan sesudah z-score adalah sama. Hal tersebut menandakan data df (numerical) tidak ada outlier dan dapat digunakan untuk pemodelan

Hal ini sesuai dengan dengan pengamatan pada boxplot di stage 1, terlihat tidak ada outlier pada boxplot.

## E. Feature Encoding

- `Married/Single` dan `Car_Ownership` menggunakan label encoding.
- `House_Ownership` menggunakan one hot encoding
- `Profession` , `City_new` , dan `State_new` menggunakan Frequency Encoding.

- **Label Encoding**

```python
df.columns = df.columns.str.lower()

mapping_married_single = {
    'single' : 0,
    'married' : 1
}

mapping_car_ownership = {
    'no' : 0,
    'yes' : 1
}

df['marital_status'] = df['married/single'].map(mapping_married_single)
df['car_ownership'] = df['car_ownership'].map(mapping_car_ownership)
```

- **One-Hot Encoding**

```python
df_onehotencoded = pd.get_dummies(df['house_ownership'], prefix='house')
df = df.join(df_onehotencoded)
df = df.drop(columns=['house_ownership'])
```

- **Frequency Encoding**

```python
profession_freq_enc = (df.groupby('profession').size()) / len(df)
city_freq_enc = (df.groupby('city_new').size()) / len(df)
state_freq_enc = (df.groupby('state_new').size()) / len(df)
```

```python
df['profession_freq_enc'] = df['profession'].apply(lambda x : profession_freq_enc[x])
df['city_freq_enc'] = df['city_new'].apply(lambda x : city_freq_enc[x])
df['state_freq_enc'] = df['state_new'].apply(lambda x : state_freq_enc[x])
```

## F. Handle Class Imbalance

```python
# pembuatan binary label target
df['risk_class'] = df['risk_flag'] > 0.8
df['risk_class'].value_counts()
```

```python
# pemisahan features vs target
X = df[[col for col in df.columns if (str(df[col].dtype) != 'object') and col not in ['risk_class', 'risk_flag']]]
y = df['risk_class'].values
print(X.shape)
print(y.shape)
```

```python
from imblearn import under_sampling, over_sampling
X_under, y_under = under_sampling.RandomUnderSampler(0.5).fit_resample(X, y)
X_over, y_over = over_sampling.RandomOverSampler(0.5).fit_resample(X, y)
X_over_SMOTE, y_over_SMOTE = over_sampling.SMOTE(0.5).fit_resample(X, y)
```

```python
print('Original')
print(pd.Series(y).value_counts())
print('UNDERSAMPLING')
print(pd.Series(y_under).value_counts())
print('OVERSAMPLING')
print(pd.Series(y_over).value_counts())
print('SMOTE')
print(pd.Series(y_over_SMOTE).value_counts())
```

## C. Feature Tambahan

1. Gender (Male/Female) : Jenis kelamin perempuan lebih berisiko untuk gagal bayar, kemungkinan karena kurang berpengalaman dalam mengelola uang dan bisnis. Hal tersebut berkaitan dengan mayoritas beranggapan bahwa perempuan bertanggung jawab untuk mengurus anak dan memenuhi kebutuhan keluarga di rumah sedangkan laki-laki berkewajiban untuk bekerja.

https://ojk.go.id/id/data-dan-statistik/research/working-paper/Documents/WP-18-03r.pdf

2. Education (High School / Diploma/ Undergraduate Degree / Master Degree/Doctoral Degree) : Pendidikan dapat memperlihatkan kualitas peminjam. Semakin tinggi tingkat pendidikan semakin tinggi pula kemungkinan untuk mengembalikan pinjaman.

https://ojk.go.id/id/data-dan-statistik/research/working-paper/Documents/WP-18-03r.pdf , https://libertystreeteconomics.newyorkfed.org/2017/11/who-is-more-likely-to-default-on-student-loans/ 

3. Amount of Loan (Total loan amount) : Semakin besar jumlah pinjaman yang dipinjam, semakin besar kemungkinan peminjam gagal bayar.

https://www.inderscienceonline.com/doi/pdf/10.1504/EMJM.2019.099956  

4. Period of Loan (Number of days from the loan granted until
the date of maturity) : Peminjam akan dikenakan tingkat bunga yang lebih tinggi untuk pinjaman dengan jangka waktu yang lebih lama. Biasanya pemberi pinjaman memilih untuk meminjamkan dengan periode waktu yang pendek untuk meminimalisir risiko gagal bayar.

https://ojk.go.id/id/data-dan-statistik/research/working-paper/Documents/WP-18-03r.pdf

5. Frequency of Loan (Number of times clients receive loan facility ) : Semakin sering customer melakukan pinjaman maka lebih banyak akses customer untuk mendapat fasilitas pinjaman, maka akan semakin tinggi kemungkinan gagal bayar.

https://www.inderscienceonline.com/doi/pdf/10.1504/EMJM.2019.099956
